```python
from pycaret.datasets import get_data
from pycaret.classification import setup
```

```python
df = get_data(dataset="diabetes")
```

```python
df.columns
```

```python
?setup
```

```python
model_setup = setup(data=df, target="Class variable")
```

```python
best_model = model_setup.compare_models()
```

```python
model_setup.predict_model(best_model)
```

```python
output_df = model_setup.predict_model(best_model, data=df)
```

```python
pipeline, model_path = model_setup.save_model(best_model, "coolModel")
```
