```python
#Import bibliotek
import pandas as pd
from pycaret.regression import *
```

```python
#Pobranie danych i zapisanie do dataframe
data_frame = pd.read_csv("../data/01_raw/car_prices_small.csv", error_bad_lines=False)
data_frame['saledate'] = pd.to_datetime(data_frame['saledate'], utc=True)
data_frame.dtypes
```

```python
model_setup = setup(data=data_frame, target="sellingprice")
```

```python
#Porównanie modeli
best_model =compare_models()
```

```python
#Ewaluacja najlepszego modelu
evaluate_model(best_model)
```

```python
output_df = predict_model(best_model, data=data_frame)
```
